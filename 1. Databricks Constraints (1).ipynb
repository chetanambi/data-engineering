# Constraints

## 1. Add constraints

```sql
CREATE OR REPLACE TABLE orders (
    order_id        BIGINT       NOT NULL,
    customer_id     BIGINT       NOT NULL,
    order_date      DATE         NOT NULL,
    order_status    STRING       NOT NULL,
    order_amount    DECIMAL(12,2) NOT NULL,
    payment_mode    STRING       NOT NULL
);

ALTER TABLE orders ALTER COLUMN customer_id SET NOT NULL;
```

```sql
ALTER TABLE orders 
ADD CONSTRAINT chk_order_status
CHECK (
    order_status IN ('CREATED', 'CONFIRMED', 'SHIPPED', 'DELIVERED','CANCELLED') 
);

ALTER TABLE orders
ADD CONSTRAINT chk_payment_mode
CHECK (
    payment_mode IN ('CASH', 'CARD', 'UPI', 'NET_BANKING')
);
```

```sql
DESCRIBE DETAIL orders;
```

```sql
SHOW TBLPROPERTIES orders;
```

```sql
INSERT INTO orders VALUES
(1001, 201, DATE '2025-02-01', 'CREATED',   1500.00, 'UPI'),
(1002, 202, DATE '2025-02-02', 'SHIPPED',   3200.50, 'CARD'),
(1003, 203, DATE '2025-02-03', 'DELIVERED',  899.99, 'NET_BANKING');
```

```sql
SELECT * FROM orders;
```

## 2. Primary Key and Foreign Key Constraints

**_Primary key and foreign key constraints require Unity Catalog and Delta Lake._**

```sql
CREATE OR REPLACE TABLE customers (
    customer_id   BIGINT NOT NULL,
    customer_name STRING,
    country       STRING,
    signup_date   DATE,

    CONSTRAINT pk_customers PRIMARY KEY (customer_id)
);
```

```sql
CREATE OR REPLACE TABLE orders (
    order_id        BIGINT       NOT NULL,
    customer_id     BIGINT       NOT NULL,
    order_date      DATE         NOT NULL,
    order_status    STRING       NOT NULL,
    order_amount    DECIMAL(12,2) NOT NULL,
    payment_mode    STRING       NOT NULL,

    CONSTRAINT pk_orders PRIMARY KEY (order_id),
    CONSTRAINT fk_orders_customers FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
```

```sql
DESCRIBE TABLE EXTENDED orders
```

```sql
INSERT INTO customers VALUES
(1, 'Chetan', 'India', '2024-01-01'),
(1, 'Ambi', 'India', '2024-01-01');

SELECT * FROM customers;
```

```sql
INSERT INTO orders VALUES
(1001, 1, DATE '2025-02-01', 'CREATED', 100.00, 'UPI'),
(1001, 999, DATE '2025-02-02', 'CREATED', 500.00, 'UPI');

SELECT * FROM orders;
```

**PK and FK are informational only but they are helpful in:**

- Query optimization using primary key constraints (RELY) (helps to eliminate unnecessary aggregration and joins)
- View the Entity Relationship Diagram

## 3. Drop Constraints

```sql
CREATE OR REPLACE TABLE customers (
    customer_id   BIGINT NOT NULL,
    customer_name STRING,
    country       STRING,
    signup_date   DATE,

    CONSTRAINT pk_customers PRIMARY KEY (customer_id)
);
```

```sql
CREATE OR REPLACE TABLE orders (
    order_id        BIGINT       NOT NULL,
    customer_id     BIGINT       NOT NULL,
    order_date      DATE         NOT NULL,
    order_status    STRING       NOT NULL,
    order_amount    DECIMAL(12,2) NOT NULL,
    payment_mode    STRING       NOT NULL,

    CONSTRAINT pk_orders PRIMARY KEY (order_id),
    CONSTRAINT fk_orders_customers FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);

ALTER TABLE orders 
ADD CONSTRAINT chk_order_status
CHECK (
    order_status IN ('CREATED', 'CONFIRMED', 'SHIPPED', 'DELIVERED','CANCELLED') 
);
```

```sql
ALTER TABLE orders DROP CONSTRAINT chk_order_status
```

```sql
ALTER TABLE orders DROP CONSTRAINT pk_orders;
```

```sql
ALTER TABLE orders DROP CONSTRAINT fk_orders_customers;
```

```sql
ALTER TABLE orders ALTER COLUMN order_id DROP NOT NULL;
```

```sql
DESCRIBE TABLE EXTENDED orders;
```
